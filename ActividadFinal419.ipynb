Un caudal de 500 m3/s de aire a 50 °C y una tempera¬tura de bulbo húmedo de 30 °C se mezcla con aire exterior que está a 15 °C y al 50 % de humedad relativa. La mezcla resultante tiene una temperatura de 25 °C. Calcula la humedad absoluta de la mezcla resultante, el caudal de aire exterior y el caudal de mezcla resultante.

```python
#Importación de módulos
import os
import psychrolib
from psychrochart import PsychroChart
import matplotlib.pyplot as plt

#Variables para los módulos
P=103025

#Sistema de unidades
psychrolib.SetUnitSystem(psychrolib.SI)

#Datos iniciales

q1=500
Ts1=50
Th1=30
W1=psychrolib.GetHumRatioFromTWetBulb(Ts1,Th1,P)
Hr1=psychrolib.GetRelHumFromTWetBulb(Ts1,Th1,P)*100
Ts2=15
Hr2=25
W2=psychrolib.GetHumRatioFromRelHum(Ts2,Hr2/100,P)
Ts3=25
W3=(((Ts3-Ts1)/(Ts2-Ts1))*(W2-W1))+W1
Hr3=psychrolib.GetRelHumFromHumRatio(Ts3,W3,P)*100


q3=q1/((Ts3-Ts2)/(Ts1-Ts2))
q2=q3-q1

print(f"Humedad absoluta resultante: {round(W3*1000,2)} g/Kg.\n")
print(f"Caudal de aire exterior : {round(q2,2)} m³/s.\n")
print(f"Caudal de la mezcla resultante : {round(q3,2)} m³/s.\n")
```
