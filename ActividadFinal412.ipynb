A una caja de mezclas de un equipo de acondicionamiento de aire, llegan dos caudales de aire, uno a 20 °C y al 60 % de humedad relativa y otro a 18 °C y al 55 % de humedad relativa. Sabiendo que los caudales son de 700 m³/h y de 1500 m³/h respectivamente. Determina los siguientes valores para la mezcla resultante: 
a) temperatura seca.
b) temperatura húmeda.
c) humedad absoluta.
d) humedad relativa.

```python
from psychrolib import *
from psychrochart import PsychroChart
import matplotlib.pyplot as plt
import os

#Variables para los módulos
P=103025

#Sistema de unidades
SetUnitSystem(SI)

#Datos iniciales

Ts1=20
Hr1=60
q1=700
W1=GetHumRatioFromRelHum(Ts1,Hr1/100,P)

Ts2=18
Hr2=55
q2=1500
W2=GetHumRatioFromRelHum(Ts2,Hr2/100,P)

#Obtención de las variables del aire mezclado.

q3=q1+q2
Ts3= (q1/q3)*(Ts1-Ts2)+Ts2

W3=(((Ts3-Ts1)/(Ts2-Ts1))*(W2-W1))+W1

Hr3 =GetRelHumFromHumRatio(Ts3,W3,P)

Th3=GetTWetBulbFromRelHum(Ts3,Hr3,P)
```

El caudal resultante sería:

$$ Q_3 = Q_1 + Q_2 $$

La temperatura seca de la mezcla se obtiene de la siguiente expresión:


$$Ts_3 = (Q_1/Q_3)*(Ts_1-Ts_2)+Ts_2$$

La humedad absoluta se obtiene utilizando la siguiente expresión:

$$W_3=\frac{(Ts_3-Ts_1)}{(Ts_2-Ts_1)}*(W_2-W_1)+W_1$$

El resultado es el siguiente:

```python
print(f"Caudal de mezcla = {q3} m³/h.")
print(f"Temperatura seca de la mezcla = {round(Ts3,2)} ªC.")
print(f"Temperatura húmeda de la mezcla = {round(Th3,2)} ªC.")
print(f"Humedad absoluta de la mezcla = {round(W3*1000,2)} g/Kg.")
print(f"Humedad relativa de la mezcla = {round(Hr3*100,2)} %.")
```
