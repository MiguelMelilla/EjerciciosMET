Un flujo de 10 m3/s de aíre húmedo a 30 °C y al 80 % de humedad relativa se calienta hasta 50 °C de tal forma que, durante su calentamiento, la humedad absoluta permanece constante. Usa el diagrama psicrométrico y calcula la cantidad de energía térmica requerida para efectuar el calentamiento.

```python
#Importación de módulos
import os
import psychrolib
from psychrochart import PsychroChart
import matplotlib.pyplot as plt

#Variables para los módulos
P=103025

#Sistema de unidades
psychrolib.SetUnitSystem(psychrolib.SI)

#Datos iniciales

Ts1=30
Hr1=80

q1=10
W1=psychrolib.GetHumRatioFromRelHum(Ts1,Hr1/100,P)
Ts2=50
W2=W1
Hr2=psychrolib.GetRelHumFromHumRatio(Ts2,W2,P)*100

E1=psychrolib.GetDryAirEnthalpy(Ts1)
E2=psychrolib.GetDryAirEnthalpy(Ts2)
print(f"Entalpía punto 1 : {round(E1/1000,2)} KJ/Kg\n")
print(f"Entalpía punto 2 : {round(E2/1000,2)} KJ/Kg\n")
print(f"Cantidad de energía : {round((E2-E1)/1000,2)} KJ/Kg\n")
```
