Una UTA trata el aire de una sala de reuniones de un edificio de oficinas. Se conocen los siguientes datos:
			Condiciones de la sala: Ts = 17ºC y Hr = 90 %
		Caudal de aire de retorno que entra en la UTA: 40 m³/h con una Ts = 27 ºC y Th = 18 ºC.
	Caudal de aire exterior que entra al climatizador: 15 m³/h con una Ts = 34 ºC y Th = 24 ºC. 
	Se pide: 
		a) Calcular el punto de mezcla de aire resultante y el punto de salida de la UTA. 
		b) Identificar qué tipos de procesos se producen. 
		c) Citar las propiedades del aire que se encuentra en el local.

```python
#Importamos las clases
from psychrochart import PsychroChart
from psychrolib import *
import matplotlib.pyplot as plt
import os

#Establecemos el sistema de unidades y la presión del aire
SetUnitSystem(SI)
P= 101325

#Datos del ejercicio:

Tsretorno=27
Thretorno=18
qretorno=40

Tsexterior=34
Thexterior=24
qexterior=15

#Obtenemos las humedades absolutas y las relativas

Hrretorno= GetRelHumFromTWetBulb(Tsretorno, Thretorno, P)
Hrexterior=GetRelHumFromTWetBulb(Tsexterior, Thexterior, P)
Wretorno= GetHumRatioFromRelHum(Tsretorno,Hrretorno,P)
Wexterior= GetHumRatioFromRelHum(Tsexterior,Hrexterior,P)

#Obtención de las variables del aire mezclado.

qmezcla=qretorno+qexterior
Tsmezcla= (qretorno/qmezcla)*(Tsretorno-Tsexterior)+Tsexterior
Wmezcla=(((Tsmezcla-Tsretorno)/(Tsexterior-Tsretorno))*(Wexterior-Wretorno))+Wretorno
Hrmezcla =GetRelHumFromHumRatio(Tsmezcla,Wmezcla,P)
Thmezcla=GetTWetBulbFromRelHum(Tsmezcla,Hrmezcla,P)

print(f"Wmezcla: {round(Wmezcla*1000,2)} g/Kg")
print(f"Wretorno: {round(Wretorno*1000,2)} g/Kg")
print(f"Wexterior: {round(Wexterior*1000,2)} g/Kg")

print(f"Tsmezcla: {round(Tsmezcla,2)} ºC")
print(f"Tsretorno: {round(Tsretorno,2)} ºC")
print(f"Tsexterior: {round(Tsexterior,2)} ºC")

print(f"Hrmezcla: {round(Hrmezcla*100,2)} %")
print(f"Hrretorno: {round(Hrretorno*100,2)} %")
print(f"Hrexterior: {round(Hrexterior*100,2)} %")
```
